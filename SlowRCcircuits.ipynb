```python
import numpy as np
import matplotlib.pyplot as plt


V = np.array([0, 4.311, 7, 8.55, 9.57, 10.21, 10.61, 10.87, 11.04, 11.15, 11.22, 11.27, 11.3])

t = np.arange(0, 130, 10)

plot_title= 'Voltage over Time for a Charging RC Circuit'
x_label= 'Time (s)'
y_label= 'Voltage (V)'

plt.plot(t,V)
plt.xlabel(x_label)
plt.ylabel(y_label)
plt.title(plot_title);
```
